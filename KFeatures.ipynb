```python
def YCbCr(R,G,B):
    #Define thresholds in YCbCr color space
    y_min = 80
    cb_min = 77
    cb_max = 135
    cr_min = 133
    cr_max = 185

    #Transform R G B in Y Cb Cr
    y = 0.299*R + 0.587*G + 0.114*B
    cr = R - y + 128
    cb = B - y + 128
    img= (y>y_min)&(cb>cb_min)&(cb<cb_max)&(cr>cr_min)&(cr<cr_max)
    return img

def ABG(R,G,B):
    #Normalize R G B channels
    r = np.divide(R,(R+G+B))
    g = np.divide(G,(R+G+B))
    b = np.divide(B,(R+G+B))
    #Define alpha beta and gamma for r g b thresholding
    alpha = (3*b*(r**2))/((r+g+b)**3)
    beta = (r+g+b)/(3*r) + (r-g)/(r+g+b)
    gamma = (r*b+(g**2))/(g*b)
    #Define thresholds in r g b color space
    alpha0 = 0.1276
    beta0 = 0.9498
    gamma0 = 2.7775
    img= (alpha>alpha0)&(beta<=beta0)&(gamma<=gamma0)
    return img

class SkinImage():
    def __init__(self,image):
        self.img= image
        R= image[:,:,0].astype("f")
        B=image[:,:,1].astype("f")
        G=image[:,:,2].astype("f")
        self.abg= ABG(R,B,G)
        self.ycc= YCbCr(R,B,G)
```

```python
from scipy import fftpack
image=obj.img[:,:,1]
# Take the fourier transform of the image.
F1 = fftpack.fft2(image)

# Now shift the quadrants around so that low spatial frequencies are in
# the center of the 2D fourier transformed image.
F2 = fftpack.fftshift( F1 )

# Calculate a 2D power spectrum
ps2D = np.abs( F2 )**2

gauss = scipy.ndimage.gaussian_filter(np.log10(ps2D),2)

f,ax= plt.subplots(1,3)
ax[0].imshow(image,cmap="gray")
ax[1].imshow(np.log10(ps2D))
ax[2].imshow(gauss)
plt.show()

```

```python
def myfft(image,filt):
    '''apply filter "filt" to fourier transorm of image and return real part of inv fft of that'''
    ft = fftpack.fft2(image)
    ft_shift= fftpack.fftshift(ft)
    ft_filt= ft_shift*filt
    shift_back= fftpack.ifftshift(ft_filt)
    inv =fftpack.ifft2(shift_back)
    return inv.real

f,axis= plt.subplots(2,4)
ax=axis.flatten()
ax[0].imshow( myfft(image,Filter.loPass),cmap="gray")
ax[1].imshow( myfft(image,Filter.hiPass),cmap="gray")
ax[2].imshow( myfft(image,Filter.horiz),cmap="gray")
ax[3].imshow( myfft(image,Filter.vert),cmap="gray")
ax[4].imshow( myfft(image,Filter.ihoriz),cmap="gray")
ax[5].imshow( myfft(image,Filter.ivert),cmap="gray")
ax[6].imshow( myfft(image,inCont),cmap="gray")
ax[7].imshow( myfft(image,outCont),cmap="gray")
```
