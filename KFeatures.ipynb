```python
def YCbCr(R,G,B):
    #Define thresholds in YCbCr color space
    y_min = 80
    cb_min = 77
    cb_max = 135
    cr_min = 133
    cr_max = 185

    #Transform R G B in Y Cb Cr
    y = 0.299*R + 0.587*G + 0.114*B
    cr = R - y + 128
    cb = B - y + 128
    img= (y>y_min)&(cb>cb_min)&(cb<cb_max)&(cr>cr_min)&(cr<cr_max)
    return img

def ABG(R,G,B):
    #Normalize R G B channels
    r = np.divide(R,(R+G+B))
    g = np.divide(G,(R+G+B))
    b = np.divide(B,(R+G+B))
    #Define alpha beta and gamma for r g b thresholding
    alpha = (3*b*(r**2))/((r+g+b)**3)
    beta = (r+g+b)/(3*r) + (r-g)/(r+g+b)
    gamma = (r*b+(g**2))/(g*b)
    #Define thresholds in r g b color space
    alpha0 = 0.1276
    beta0 = 0.9498
    gamma0 = 2.7775
    img= (alpha>alpha0)&(beta<=beta0)&(gamma<=gamma0)
    return img

class SkinImage():
    def __init__(self,image):
        self.img= image
        R= image[:,:,0].astype("f")
        B=image[:,:,1].astype("f")
        G=image[:,:,2].astype("f")
        self.abg= ABG(R,B,G)
        self.ycc= YCbCr(R,B,G)
```

```python
from scipy import fftpack
image=obj.img[:,:,1]
# Take the fourier transform of the image.
F1 = fftpack.fft2(image)

# Now shift the quadrants around so that low spatial frequencies are in
# the center of the 2D fourier transformed image.
F2 = fftpack.fftshift( F1 )

# Calculate a 2D power spectrum
ps2D = np.abs( F2 )**2

f,ax= plt.subplots(1,2)
ax[1].imshow(np.log10(ps2D))
ax[0].imshow(image)
plt.show()

```
